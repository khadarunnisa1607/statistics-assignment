1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

ans:-Covariance – Definition

-Covariance is a statistical measure that shows the direction of the relationship between two variables.

-If both variables increase or decrease together → positive covariance.

-If one increases while the other decreases → negative covariance.

-If there is no consistent relationship → covariance is close to zero.

Correlation is a standardized version of covariance that measures both direction and strength of the relationship between two variables.

Values range from –1 to +1

+1 → perfect positive relationship

–1 → perfect negative relationship

0 → no relationship



2)What does a positive, negative, and zero covariance indicate about the relationship between two variables?


ANS:-
Covariance tells you how two variables move in relation to each other. The sign (positive, negative, or zero) gives the direction of this relationship:

1. Positive Covariance (Cov > 0)

Indicates that both variables move in the same direction

When one variable increases, the other also increases
When one decreases, the other also decreases

Example:
Income and expenditure — as income increases, spending usually increases

2. Negative Covariance (Cov < 0)

 Indicates that variables move in opposite directions

When one variable increases, the other decreases
When one decreases, the other increases

Example:
Price and demand — as price increases, demand typically decreases

3. Zero Covariance (Cov ≈ 0)

Indicates no linear relationship between variables

Changes in one variable do not show any consistent pattern with the other

Example:
Shoe size and intelligence — no meaningful relationship

3)Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation
preferred in many cases?


Limitations of Covariance

Covariance is useful for identifying the direction of a relationship, but it has several important limitations:

1. Depends on Scale (Units Problem)
Covariance values change if the units of variables change (e.g., meters → centimeters)
This makes it inconsistent and difficult to compare
2. No Fixed Range
Covariance can take any value from –∞ to +∞
Because of this, you cannot judge strength (whether relationship is strong or weak)
3. Difficult Interpretation
A value like 500 or –200 does not clearly indicate how strong the relationship is
It only shows direction (positive/negative), not magnitude meaningfully
4. Not Comparable Across Datasets
Since it depends on scale, covariance values from different datasets cannot be compared directly
5. Affected by Outliers
Extreme values can distort covariance, leading to misleading results
6. Only Measures Linear Relationship
Covariance only captures linear relationships and may miss other patterns
Why Correlation is Preferred

Correlation overcomes most of these limitations:

1. Standardized Measure
Correlation values always lie between –1 and +1
Makes it easy to interpret and compare
2. Unit-Free (Dimensionless)
Not affected by units of measurement
Provides consistent results across datasets
3. Shows Strength + Direction
Clearly indicates both:
Direction (positive/negative)
Strength (weak/moderate/strong)
4. Easy Comparison
You can compare relationships between different variables or datasets easily
5. Better Interpretation
Example:
Covariance = 300 → unclear
Correlation = 0.9 → strong positive relationship ✔
Conclusion

Covariance is limited because it is scale-dependent and hard to interpret, whereas correlation is standardized, unit-free, and more meaningful, making it the preferred measure for analyzing relationships between variables.


4)Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation
coefficient. When would you prefer to use Spearman’s correlation?


ANS:-pearson’s vs Spearman’s Correlation Coefficient

Both are used to measure the relationship between two variables—but they differ in what they measure and when they are used.

1. Pearson’s Correlation Coefficient
Measures linear relationship between two variables
Uses actual numerical values
Assumes data is:
Normally distributed
Linearly related
Sensitive to outliers

 Example: Relationship between height and weight

2. Spearman’s Rank Correlation Coefficient
Measures monotonic relationship (not necessarily linear)
Uses ranks instead of actual values
Does not require normal distribution
Less affected by outliers

Example: Ranking students in two subjects

5)If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you
infer causation from this value? Why or why not?

ANS:-nterpretation of Correlation Coefficient (r = 0.85)

A correlation coefficient of 0.85 indicates a strong positive relationship between variables X and Y.

As X increases, Y also tends to increase
The relationship is strong (since it is close to +1)
The data points would lie close to a straight upward-sloping line

In simple terms: X and Y move together in a strong and consistent way

Can We Infer Causation?  No

A correlation of 0.85 does NOT imply causation.

Why?
1. Correlation ≠ Causation
Just because two variables are related does not mean one causes the other
2. Possibility of Third Variable (Hidden Factor)
A third variable (called a lurking variable) may influence both X and Y

Example:
Ice cream sales and sunglasses sales may both increase due to hot weather

3. Direction of Influence is Unknown
Even if there is a relationship, we don’t know:
Does X cause Y?
Or does Y cause X?
4. Coincidental Relationship
Sometimes variables move together purely by chance
Conclusion
r = 0.85 → strong positive correlation ✔
Causation cannot be concluded
Further analysis (like experiments or causal models) is required to establish cause-and-effect
