4.The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?



ANS:-Here we use stratified sampling.
Because;

The population has two groups (strata):

Boys = 60%

Girls = 40%

In Stratified Sampling, the population is divided into groups (strata) based on characteristics such as gender, and then samples are taken from each group separately.

If you want equal representation of both genders, you can select the same number of boys and girls from each group.
