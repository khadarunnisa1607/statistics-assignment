4.What is the concept of standardization (z-score), and why is it important in data analysis? Explain the
formula and how standardization transforms a dataset.



ANS:-

Concept of Standardization (Z-Score)

Standardization is a method used to convert data values into a standard scale so that they can be compared easily.
This standardized value is called the z-score.

A z-score tells us how many standard deviations a value is away from the mean of the dataset.

Formula

𝑧=X-μ​/σ

Where:

z = z-score (standardized value)

X = original data value

μ (mu) = mean of the dataset

σ (sigma) = standard deviation of the dataset.

How Standardization Transforms a Dataset

When data is standardized:

The mean becomes 0

The standard deviation becomes 1

Values show their distance from the mean in standard deviation units

Example:
If a student scored 60 and the class mean is 50 with standard deviation 5:
Z=60-50/5=2

This means the score is 2 standard deviations above the mean.


Why Standardization is Important in Data Analysis

1.Comparison of different datasets
It allows comparison between values measured on different scales.

2.Identifying outliers
Extremely high or low z-scores indicate unusual values.

3.Useful in statistical models
Many algorithms like regression, clustering, and machine learning models work better with standardized data.

4.Understanding data position
It shows whether a value is above or below the average.

Standardization converts data into a common scale where the mean = 0 and standard deviation = 1, making analysis and comparison easier.

5.What is Kurtosis and their type?

ANS:-

Kurtosis is a statistical measure that describes the shape of a distribution, especially the peakedness and heaviness of the tails compared to a normal distribution.

In simple terms, it tells us how sharp the peak is and how heavy the tails are in a dataset.


Types of Kurtosis
1️ Mesokurtic

Distribution similar to a normal distribution

Moderate peak and tails

Kurtosis value ≈ 3

Example: Normal distribution


2.Leptokurtic

Distribution has a very sharp peak

Heavy tails (more extreme values or outliers)

Kurtosis value greater than 3

Meaning: Data is more concentrated around the mean but has more outliers.


3️.Platykurtic

Distribution has a flatter peak

Light tails (fewer outliers)

Kurtosis value less than 3

Meaning: Data is more spread out.


Kurtosis helps us understand how peaked or flat a distribution is compared to a normal distribution.


6.Explain why the uniform distribution is a good model for the outcome of rolling a fair die.


ANS:-

Why Uniform Distribution Models a Fair Die

A uniform distribution occurs when all possible outcomes have equal probability.

When a fair die is rolled, the possible outcomes are:

1, 2, 3, 4, 5, 6

Each outcome has the same probability:
p(1)=p(2)=p(3)=p(4)=P(5)=p(6)=1/6


Reasons it fits a Uniform Distribution

Equal Probability
Every number from 1 to 6 has an equal chance of occurring.

Discrete Outcomes
The die has a finite number of outcomes, and each outcome is equally likely.

No Bias
A fair die means no side is favored, so the probabilities remain the same.

Conclusion

The outcome of rolling a fair die follows a discrete uniform distribution because each possible result (1–6) occurs with equal probability of 1/6.

8.How does log transformation help in stabilizing variance and making data more normally distributed?

ANS:-

Log Transformation

Log transformation is a technique used in data analysis where the logarithm of each data value is taken (e.g., log(X)). It is commonly used when data is highly skewed or has large variability.


1. Stabilizing Variance

In many datasets, variance increases with the size of the values. Large values may vary much more than small values.

Log transformation helps by:

Reducing the spread of large values

Compressing the scale of the data

Making variability more consistent across the dataset

Example:
Values like 10, 100, 1000 have large gaps, but after log transformation:

log(10) = 1

log(100) = 2

log(1000) = 3

The differences become smaller and more manageable, stabilizing the variance.


2. Making Data More Normally Distributed

Many real-world datasets are right-skewed (long tail on the right side).

Log transformation:

Reduces extreme high values

Pulls the right tail closer to the center

Makes the distribution more symmetric and closer to a normal distribution

3. Benefits in Data Analysis

Log transformation helps because it:

Reduces skewness

Stabilizes variance

Makes data better suited for statistical tests and regression models

Helps in identifying patterns and relationships



Log transformation compresses large values, reduces skewness, and stabilizes variance, making the dataset closer to a normal distribution and easier to analyze.
